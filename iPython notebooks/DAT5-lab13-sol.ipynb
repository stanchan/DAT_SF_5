#### c. indexing and assignment

Let's look at indexing into our data:

    >>> c
    
    >>> c[1][1]

    >>> d[1][1]

Assigning new values - part 1:

    >>> c[1][1] = 7
    
    # what's the problem here?
    
    >>> d[1][1] = 7
    
    >>> e[1][1] = 7
    
    >>> c[1]
    
    >>> c[1] = (2, 7)
    
    # works if we assign per-tuple, rather than tuple-element.
    
Assigning new values - part 2:
    
    >>> c[1][1]
    
    >>> c[1][1] = 'seven'
    
    >>> c[1] = (2, 'seven')
    
    >>> c[1][1]
    
    >>> d[1][1]
    
    >>> d[1][1] = 'seven'
    
    # no strings in a numpy int array
    
    >>> e[1][1] = 'seven'
    
Assigning new values - part 3:

    >>> c[1] = (2, 7.1)
    
    >>> e[1][1] = 7.1
    
    >>> d[1][1] = 7.1
        
    >>> d

    # float gets converted to int!  You can use np.dtype to check the type:
    
    >>> d.dtype            # shows the type of data in the np array
    
    # convert to float

    >>> d.astype(float)
    
    >>> d[1][1] = 7.1

    >>> d

    # our astype returned the float array, but did not alter 'in place'
    
    >>> d = d.astype(float)

    >>> d

    >>> d.astype(int)

#### d. values, references, and copies

Some new variables:

    >>> x = 1

    >>> y = 2
    
    >>> z = 3
    
    >>> a = [x, y, z]
    
    >>> a
    
    >>> b = a
    
    >>> b
    
Assignment alone just creates new names pointing to existing elements!

    >>> a[0] = 4
    
    >>> a
    
    >>> b
    
    # values of b changed as well - and it works both ways:
    
    >>> b[0] = 5
    
    >>> b
    
    >>> a
    
Notice that the original x, y, z were not 'linked' - they were evaulated at the time a was created, but were then separate.

    >>> x
    
To create a new list, rather than a new reference to existing list, use the list constructor explicitly:

    >>> b = list(a)
    
    >>> a[0] = 4
    
    >>> a
    
    >>> b

This works for simple (i.e. flat) iterables, but what about more complex, nested objects?

    >>> a = [1, 2, 3]
    
    >>> b = [4, 5, 6]
    
    >>> c = [a, b]
    
    >>> c
    
    >>> a[0] = 7
    
    >>> c
    
    >>> c = list([a, b])
    
    >>> a[0] = 1
    
    >>> c
    
Internal objects are still references, even though top level is copy!  To get around this problem we have to use deep copies (i.e. everything is a copy, not references!)

    >>> import copy
    
    >>> d = copy.deepcopy(c)
    
    >>> a[0] = 70
    
    >>> c
    
    >>> d
    
Summary:  Double-check new assignments, references, and copies to make sure you're getting the expected behavior.
